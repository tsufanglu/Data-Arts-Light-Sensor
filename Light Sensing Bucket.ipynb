```python
import csv
import datetime
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
FILE_PATH = "Light Sensing Bucket.csv"
max_data, ts, min_data, avg_data = [],[],[],[]

with open(FILE_PATH,'rw') as f:
    reader = csv.reader(f)
    for line in reader:
        time = line[0].replace("Z","").replace("T"," ")[0:19]
        time = datetime.datetime.strptime(time,"%Y-%m-%d %H:%M:%S")
        if time not in ts:
            ts.append(time)
        data_type, data_point = line[1], line[2]
        if data_type == "max":
            max_data.append(data_point)
        elif data_type == "min":
            min_data.append(data_point)
        else:
            avg_data.append(data_point)
    ts = np.array(list(ts))
plt.plot(ts,max_data,label='max')
plt.plot(ts,min_data,label='min')
plt.plot(ts,avg_data,label='avg')
plt.xlabel("Time")
plt.ylabel("Sensor Reading")
plt.xticks(rotation='vertical')
plt.legend(loc="upper right")
```
